# 1. Replacing NaNs with mean for highway miles per gallon variable by using Simple Imputer method.

```python
cars = pd.read_excel('C:/Users/DostTech/Desktop/DSA/Python Programming/cars.xlsx')
cars
```

```python
cars.isna()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
imputer=imputer.fit(cars[['hwy']])
cars['hwy'] = imputer.transform(cars[['hwy']]).ravel()
```

```python
cars.isnull().sum()
```

# 2. Replacing NaNs with backfill method for engine displacement and city miles per gallon variables.

```python
cars['displ']=cars['displ'].fillna(method='backfill')
```

```python
cars['cty']=cars['cty'].fillna(method='backfill')
```
